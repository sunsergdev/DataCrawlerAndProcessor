```python
import os
import json

cwd = os.getcwd()
birdnest_dir = os.path.join(cwd, "birdnest")
articles_dir = os.path.join(birdnest_dir, "articles")
article_files = [os.path.join(articles_dir, file) for file in os.listdir(articles_dir)]
images_dir = os.path.join(birdnest_dir, "images")
if not os.path.exists(images_dir):
  os.mkdir(images_dir)
  print("Folder %s created!" % images_dir)
else:
  print("Folder %s already exists" % images_dir)
```

```python
import urllib

from concurrent.futures import ThreadPoolExecutor

def fetch_images(file): 
    with open(file) as f:
        data = json.loads(f.read())
        images = list(map(lambda image: "https://www.birdsnest.com.au"+image["data"].replace(":", "").format(size = "1280x1080"), data["product"]["images"]))
        product_img_dir = os.path.join(images_dir, os.path.splitext(os.path.basename(f.name))[0])
        if not os.path.exists(product_img_dir):
            os.mkdir(product_img_dir)
            print("Folder %s created!" % product_img_dir)
            for (idx, image) in enumerate(images):
                urllib.request.urlretrieve(image, os.path.join(product_img_dir, "{idx}.jpg".format(idx=idx)))
        else:
            print("Folder %s already exists" % product_img_dir)
with ThreadPoolExecutor(100) as ex:
    for file in article_files:
        ex.submit(fetch_images, file)
```
